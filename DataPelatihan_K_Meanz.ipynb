```python
# memasukkan file dataset dan membaca csv
import pandas as pd
df = pd.read_csv('/content/Dataset Pelatihan.csv')

df.head(10)
```

```python
# mengubah nama kolom menjadi huruf kapital (uppercase)
df.rename(columns=str.upper, inplace=True)

df.head()
```

```python
# mengubah data yang ada pada baris (ubah nama agus Siti menjadi Sena)
df['NAMA'] = df['NAMA'].replace('Siti','Sena')

df.head()
```

```python
#mengubah data kategorik menjadi numerik (gender & matkul)
df['GENDER'].replace(['Laki-laki', 'Perempuan'], [0,1], inplace=True)
df['MATKUL'].replace(['Matematika','Fisika','Kimia','Biologi'], [1,2,3,4], inplace=True)

df.head()
```

```python
# membersihkan angka dari NaN, menghilangkan koma yang tidak perlu, dan menjadikan integer
df['ID'].fillna(0, inplace=True)
df['UMUR'].fillna(0, inplace=True)
df['MATKUL'].fillna(0, inplace=True)
df['GENDER'].fillna(0, inplace=True)
df['NILAI'].fillna(0, inplace=True)
df['UTS'].fillna(0, inplace=True)
df['UAS'].fillna(0, inplace=True)
df['ID'] = df['ID'].astype(int)
df['UMUR'] = df['UMUR'].astype(int)
df['MATKUL'] = df['MATKUL'].astype(int)
df['GENDER'] = df['GENDER'].astype(int)


df.head()
```

```python
# menghhapus kolom NAMA, karna merupakan data kategorik
df.drop(['NAMA'], axis=1, inplace=True)

df.head(20)
```

```python
# menghapus kolom tanggal karna tidak bisa di proses dibagian inertia
df.drop(['TANGGAL'], axis=1, inplace=True)

df.head()
```

```python
#membuat list inertia
from sklearn.cluster import KMeans

X = df
clusters = []
for i in range(1,11):
  km = KMeans(n_clusters=i).fit(X)
  clusters.append(km.inertia_)
```

```python
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns

# membuat plot inertia
fig, ax = plt.subplots(figsize=(8, 4))
sns.lineplot(x=list(range(1, 11)), y=clusters, ax=ax)
ax.set_title('Cari Elbow')
ax.set_xlabel('Clusters')
ax.set_ylabel('Inertia')
```

```python
# membuat objek KMeans
km5 = KMeans(n_clusters=2).fit(X)
```

```python
# menambahkan kolom label pada dataset
X['Labels'] = km5.labels_
```

```python
# membuat plot KMeans dengan 5 klaster
plt.figure(figsize=(8,4))
sns.scatterplot(x=X['ID'], y=X['NILAI'], hue=X['Labels'],
                palette=sns.color_palette('hls', 5))
plt.title('KMeans dengan 5 Cluster')
plt.show()
```
